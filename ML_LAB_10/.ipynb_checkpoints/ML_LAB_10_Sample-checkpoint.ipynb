## Random Forest Classifier 

- Random forest is a Supervised Learning algorithm which uses ensemble learning method for classification and regression.
- It is a bagging technique and not a boosting technique. 
- The trees in random forests are run in parallel. There is no interaction between these trees while building the trees.
- The diagram of the Random Forest is given as follows:

<img src = "rfs.PNG" height = "700" width ="600">

- A forest is comprised of trees. It is said that the more trees it has, the more robust a forest is. 
- Random forests creates decision trees on randomly selected data samples, gets prediction from each tree and selects the best solution by voting.

## Ensemble Learning
An Ensemble method is a technique that combines the predictions from multiple machine learning algorithms together to make more accurate predictions than any individual model. A model comprised of many models is called an Ensemble model.

## Importing Required Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import tree
import graphviz 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
wine = pd.read_csv("winequality-red.csv")
```

```python
wine.head()
```

### Finding the Quality

```python
quality_dist = wine['quality'].value_counts()
plt.bar(quality_dist.index, quality_dist)
plt.xlabel('quality')
plt.ylabel('frequency')
plt.show()
```

- The quality of the red wines are in the set  {3,4,5,6,7,8} . 
- We want to classify the wine into so-called 'bad' and 'good'. 
- We study statistics of our dataset to determine how to make a classification.

### Correlation Via HeatMap

```python
f,ax = plt.subplots(figsize = (20,20))
plt.title("Correlation Heatmap", fontsize=20)
sns.set(font_scale=1.1)
sns.heatmap(wine.corr(), linewidth = 5, linecolor = "white", annot = True, ax = ax)
plt.yticks(rotation='horizontal')
plt.show()
```

### 5 - Point Summary

```python
wine['quality'].describe()
```

### Finding the Median Quality Range for "Good" vs "Bad" Classification

```python
values, base = np.histogram(wine['quality'], bins=20)
kumulativ = np.cumsum(values/wine.shape[0])
plt.plot(base[:-1], kumulativ, c='blue')
plt.xlabel('quality')
plt.ylabel('frequency')
plt.show()
```

- Since the median quality is  6  and mean quaity is  5 or ~6.
- We classify red wine as **'poor'** if **quality is less then or equal to  6**, otherwise we **classify the red wine as 'good'.** 
- We repleace the column quality with the **Value  0  for 'POOR' Wine** and the **Value  1  for 'GOOD' Wine**.

### Classification of Good and Bad Wine Based on Quality Condition

```python
dataForClassification = wine.loc[wine['quality'] <= 6].index
condition = wine.loc[wine['quality'] > 6].index
wine.iloc[dataForClassification, wine.columns.get_loc('quality')] = 0
wine.iloc[condition, wine.columns.get_loc('quality')] = 1
```

```python
wine['quality'].value_counts()
```

```python
x = wine.drop('quality',axis=1)
y = wine['quality']
```

### Choosing 40% as Training Data

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.40, random_state = 42)
```

## Decision Tree
- We make a decision tree to try to determine the characteristics of so-called 'good' wine in our classification.

```python
# Making a decision tree with depth = 2
clfTre = tree.DecisionTreeClassifier(max_depth=2)
clfTre.fit(xTrain, yTrain)
```

```python
#Visualizing the decision tree
dot_data = tree.export_graphviz(clfTre, out_file=None, max_depth=2, feature_names=list(x.columns.values), filled=True, rounded=True)
valgTre = graphviz.Source(dot_data) 
valgTre
```

```python
utfall = (clfTre.predict(xTest) == yTest).value_counts()
print("The Decision Tree Predicts the Test Data in", (utfall[1]/(utfall[0]+utfall[1]))*100 , "% of the cases.")
```

## Random Forest Classifier

```python
rf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=0)
rf.fit(xTrain, yTrain)
```

```python
utfall = (rf.predict(xTest) == yTest).value_counts()
print("The Random Forest Predicts the Test Data in", (utfall[1]/(utfall[0]+utfall[1]))*100 , "% of the cases.")
```
